```python
import numpy as np
import pyqstrat as pq

dates = np.arange(np.datetime64('2016-01-01'), np.datetime64('2018-03-01'))
returns = np.random.normal(size = len(dates)) / 100.
metrics = pq.compute_return_metrics(dates, returns, 1.e6).metrics()
pq.display_return_metrics(metrics)
pq.plot_return_metrics(metrics)
```
